```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/data.csv')
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# **Drop unamed column**

```python
df = df.drop('Unnamed: 0', axis=1)
print(df.head())
```

# **Gender Distribution**

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x='Gender', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

From the above chart we have analyzed that number of Female in data is more as compared to number of Male.


```python
gb = df.groupby('ParentEduc').agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(gb, annot=True)
plt.show()
```

From the above chart we have concluded that the education of parent have a good impact in the course.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({ "MathScore": "mean","ReadingScore": "mean","WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(gb1, annot=True)
plt.show()
```

From the above chart we have concluded that there is no/neligible impact on student's score due to their parent marital status.

# **Distribution of ethnic group**

```python
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()

l=['group A','group B','group C','group D','group E']
plt.figure(figsize=(5,5))
plt.pie(x=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']],labels=l,autopct='%1.2f%%')
plt.title('Distribution of Ethnic Group')
plt.show()
```
